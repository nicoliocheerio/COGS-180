### Simple Distributions

```python
#plot the distribution of the 'How would you classify your AI-use persona' column
#clean column names by removing information in parentheses
survey_data['How would you classify your “AI-use personality”?'] = (
    survey_data['How would you classify your “AI-use personality”?']
    .str.split('(')
    .str[0]
    .str.strip()
)

#Reorder the values in the 'How would you classify your AI-use personality?' column by the number of responses for each category
personality_order = ["AI skeptic", "Cautious user", "Generous adopter", "Overuser / Devoted User"]
#cut out row where 'How would you classify your AI-use personality' contains Medium Use
survey_data_temp = survey_data[~survey_data['How would you classify your “AI-use personality”?'].str.contains('Medium Use', case=False, na=False)]

ai_personality_dist = px.histogram(survey_data_temp, 
                                   x='How would you classify your “AI-use personality”?', 
                                   title='Distribution of AI-Use Personalities', 
                                   width =800)

ai_personality_dist.update_layout( xaxis={'categoryorder':'array', 'categoryarray': personality_order})
ai_personality_dist.show()
```

```python
#plot the distribution of 'In this situation, using AI felt like the most rational choice.' column
rational_choice_dist = px.histogram(survey_data, 
                                   x='In this situation, using AI felt like the most rational choice.', 
                                   title='Distribution of Rational Choice Responses', 
                                   
                                   width =800)
rational_choice_dist.update_layout(bargap=0.2)

#update plot to have a label indicating the average rationality score across all responses of 3.66
average_rationality = survey_data['In this situation, using AI felt like the most rational choice.'].mean()
rational_choice_dist.add_annotation(
    x=average_rationality,
    y=0,
    text=f'Average Rationality Score: {average_rationality:.2f}',
    showarrow=True,
    ax=0,
    ay=30
)
rational_choice_dist.show()
```

```python
#find the average rationality score overall
average_rationality = survey_data['In this situation, using AI felt like the most rational choice.'].mean()
print(f'Average Rationality Score: {average_rationality:.2f}')
```

```python
#plot box and whisker plot of 'How would you classify your AI-use personality?' column by 'In this situation, using AI felt like the most rational choice.' column
box_plot = px.box(survey_data,
                  x='How would you classify your “AI-use personality”?',
                  y='In this situation, using AI felt like the most rational choice.',
                  title='Rational Choice by AI-Use Personality',
                  width=1000,
                  height = 500)

box_plot.update_layout(xaxis_title='AI-Use Personality', yaxis_title='Rationality Score', xaxis={'categoryorder':'array', 'categoryarray': personality_order})

#label median values for each box and mean values for each box offset x value by 0.2 to the right and left respectively
for personality in personality_order:
    median_value = survey_data[survey_data['How would you classify your “AI-use personality”?'] == personality]['In this situation, using AI felt like the most rational choice.'].median()
    mean_value = survey_data[survey_data['How would you classify your “AI-use personality”?'] == personality]['In this situation, using AI felt like the most rational choice.'].mean()
    box_plot.add_annotation(x=personality, y=median_value, text=f'Median: {median_value:.2f}', showarrow=False, yshift=10, xshift= 100)
    box_plot.add_annotation(x=personality, y=mean_value, text=f'Mean: {mean_value:.2f}', showarrow=False, yshift= -10, xshift= 100)


box_plot.show()
```

### Top Parameters by Spearman Correlation Per AI Personality Type

```python
#write function to filter part2_df by each of the 4 personality types and calculate spearman correlation values with rationality for each personality type
def calculate_correlations_by_personality(df, personality):
    filtered_df = df[df['How would you classify your “AI-use personality”?'] == personality]
    filtered_df = filtered_df.drop(columns=['How would you classify your “AI-use personality”?'])
    correlation_matrix = filtered_df.corr(method='spearman')
    return correlation_matrix['Rationality'].sort_values(ascending=False)

corr_generous = calculate_correlations_by_personality(part2_df, 'Generous adopter')
```

```python
#plot the distribution of 'In this situation, using AI felt like the most rational choice.' column by AI-use personality as a barplot for the top 4 most positively correlated variables with rationality for each personality type
#drop rationality from the correlation series
#corr_generous = corr_generous.drop('Rationality')

#sort the correlation values in descending order and take the top 4 most positively correlated variables with rationality

import plotly.graph_objects as go
import plotly.express as px

def create_corr_bar_plot(corr_series, personality):

    corr_series = corr_series.drop('Rationality')
    corr_series = corr_series.sort_values(ascending=False).head(5)

    colors = px.colors.sample_colorscale(
        px.colors.sequential.BuPu,
        [i/(len(corr_series)-1) for i in range(len(corr_series))]
    )

    fig = go.Figure()

    fig.add_bar(
        x=corr_series.values,
        y=corr_series.index[::-1],
        orientation='h',
        marker=dict(color=colors),
        text=corr_series.values.round(3),
        textposition='outside'
    )

    fig.update_yaxes(
    ticks="outside",
    ticklen=10,
    tickwidth=1,
    tickcolor='white',
   
)

    return fig
create_corr_bar_plot(corr_generous, 'Generous adopter')
```

```python
corr_overuser =calculate_correlations_by_personality(part2_df, 'Overuser / Devoted User')
create_corr_bar_plot(corr_overuser, 'Overuser / Devoted User')
```

```python
corr_cautious = calculate_correlations_by_personality(part2_df, 'Cautious user')
create_corr_bar_plot(corr_cautious, 'Cautious user')
```

```python
from scipy.stats import spearmanr
corr, pvals = spearmanr(part2_df)

cols = part2_df.columns

spearman_corr = pd.DataFrame(corr, index=cols, columns=cols)
pval_matrix = pd.DataFrame(pvals, index=cols, columns=cols)

# significance threshold
alpha = 0.05

# keep only statistically significant correlations
significant_corr = spearman_corr.where(pval_matrix < alpha)
```

```python
#plot significant_corr as a heatmap using plotly express
significant_heatmap = px.imshow(
    significant_corr,
    labels={"x": "Variable", "y": "Variable", "color": "Spearman Correlation"},
    x=significant_corr.columns,
    y=significant_corr.columns,
    color_continuous_scale=px.colors.sequential.BuPu,
    title="Significant Correlations Heatmap",
    width = 1000,
    height = 1000
)
significant_heatmap.show()
```

```python
#plot p values as a heatmap using plotly express 
pval_heatmap = px.imshow(
    pval_matrix,
    labels={"x": "Variable", "y": "Variable", "color": "P-value"},
    x=pval_matrix.columns,
    y=pval_matrix.columns,
    color_continuous_scale=px.colors.sequential.BuPu,
    title="P-values Heatmap",
    width = 800,
    height = 800
)
pval_heatmap.show()
```
